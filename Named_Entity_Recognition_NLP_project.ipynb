# Named Entity Recognition

### Importing data

```
# install datasets
!pip install datasets

# import amazon plarity dataset
from datasets import load_dataset
dataset = load_dataset("amazon_polarity")
```

```
print(dataset['train']['content'][0])
```

### Preparing Data

```
reviews = []

for review in dataset['train']:
  
  reviews.append(review['content']) 

print(len(reviews))
print(reviews[0])
```

### Tagging Data

```
# install flair
! pip install flair

from flair.data import Sentence
from flair.models import SequenceTagger
```

```
# load the NER tagger
tagger = SequenceTagger.load('ner')

import random

reviews_test = random.sample(reviews, 100)

for review in reviews_test:
  
  review = Sentence(review)

  # run NER over sentence
  tagger.predict(review)
  
  print(review)

  print('The following NER tags are found:')

  # iterate over entities and print
  for entity in review.get_spans('ner'):
      print(entity)

  print("-" * 100)
```
